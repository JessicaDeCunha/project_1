```python
import requests
import json
import pandas as pd
%matplotlib inline
```

```python
url = 'https://api.covid19api.com/total/country/united-states/status/confirmed'
```

```python
url = url + "?format=json"
```

```python
requests.get(url)
```

```python
response_data = requests.get(url)
```

```python
response_content = response_data.content
print(response_content)
```

```python
data = response_data.json()
print(data)
```

```python
print(json.dumps(data, indent = 4))
```

```python
cases = data[100]['Cases']
print(cases)
```

```python
case = []
date = []
for item in data:
    case.append(item['Cases'])
for item in data:
    date.append(item['Date'])
    
```

```python
case
```

```python
len(case)
daily_case = []
for i in range(len(case)-1):
    daily = case[i+1]-case[i]
    daily_case.append(daily)
daily_case
```

```python
df = pd.DataFrame(case)
df_date = pd.DataFrame(date)
df_daily = pd.DataFrame(daily_case)
df_daily_pct = df_daily.pct_change()
```

```python
joined_df = pd.concat([df, df_date], axis = 'columns', join='inner')
joined_df.columns = ['Case', 'Date']
joined_df = joined_df.set_index(pd.to_datetime(joined_df["Date"], infer_datetime_format = True))
joined_df = joined_df.drop(columns = ['Date'])
joined_df.head(20)
```

```python
daily_pct_change = joined_df.pct_change()
daily_pct_change.head(90)
```
